```python
import numpy as np
from pandas import *
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
train_index = tuple(y[0] for y in training_score)
train_data = {'score' : Series(tuple(y[1] for y in training_score), index=train_index),
 'class' : Series(tuple(y[2] for y in training_score), index=train_index)}
train_frame = DataFrame(train_data)
print train_frame.describe()

test_index = tuple(y[0] for y in testing_score)
test_data = {'score' : Series(tuple(y[1] for y in testing_score), index=test_index),
 'class' : Series(tuple(y[2] for y in testing_score), index=test_index)}
test_frame = DataFrame(test_data)
print test_frame.describe()
```

```python
def dataHistPlot(dataframe, name=""):
    mu, sigma = dataframe.mean()[1], dataframe.std()[1]
    x = []
    for y in dataframe['score']:
        x.append(1 if y > (mu+sigma) else 2 if y > (mu - sigma) else 3 )
    dataframe['ndclass'] = x
    
    x = dataframe['score'] 
    # the histogram of the data
    n, bins, patches = plt.hist(x, 20, facecolor='green', normed=True, alpha=0.75)
        
    # add a 'best fit' line
    y = mlab.normpdf( bins, mu, sigma)
    l = plt.plot(bins, y, 'r--', linewidth=1)
    
    plt.xlabel('Scores')
    plt.ylabel('Probability')
    plt.title(r'$\mathrm{Histogram\ of\ ' + name +'\ Set\ Scores:}\ \mu='+('%.2f'%mu)+',\ \sigma='+('%.2f'%sigma)+'$')
    plt.axis([0, 10, 0, 0.5])
    plt.grid(True)
    for p in patches:
        if p.get_x() > (mu+sigma):
            p.set_facecolor(cm.spring(0))
        elif p.get_x() > (mu-sigma):
            p.set_facecolor(cm.summer(0))
        else:
            p.set_facecolor(cm.winter(0))
    plt.xticks(np.arange(0, max(x)+1, 1))
    return #plt.figure(figsize=(8, 6))
```

```python
fig = plt.figure(figsize = (12,9))
plt.subplot(2,1,1)
dataHistPlot(train_frame, 'Training')
plt.subplot(2,1,2)
dataHistPlot(test_frame, 'Testing')
```
