```python
def lineSpacial( img):
    lines = img.findLines() # get the lines # return only long ones
    lineCount = len(lines)
 #   lines = lines.filter( lines.width()/img.width > .2 )
 #   lines = lines.filter( lines.height()/img.height > .2 )
    vertical = lines.filter((np.abs(lines.angle()) > 80))
    horizontal = lines.filter(np.abs(lines.angle()) < 10 )

    lineTb = pd.DataFrame( index=np.arange(len(lines)))
    lineTb.insert(0, 'w', [line.width() for line in lines])
    lineTb.insert(0, 'h', [line.height() for line in lines])
    lineTb.insert(0, 'x', [line.x for line in lines])
    lineTb.insert(0, 'y', [line.y for line in lines])
    
    mostx = 0 if len(lines) == 0 else lineTb.x.value_counts().index[0]
    mosty = 0 if len(lines) == 0 else lineTb.y.value_counts().index[0]
    #return [(lineTb.x.max()- lineTb.x.min())/img.width, (lineTb.y.max()- lineTb.y.min())/img.height]
    return [lineCount, len(vertical), len(horizontal), mostx, mosty]

def getLsFeats(imgset):
    feats = pd.DataFrame( index=['lc', 'vc', 'hc', 'mostx', 'mosty'])
    i = 0
    for img in imgset:
        #if i > 10: break
        r = lineSpacial(img)
        feats.insert(i, i+1, r)
        i = i+1
    return feats.T

def showFeats(feats, ground):
    trg = ground
    
    fig = plt.figure(figsize=(14, 13))
    fig.subplots_adjust(hspace=.3)
    
    cmap = cm.PuBu_r
    cax = fig.add_axes([0.2, 0.94, 0.6, 0.03])
    norm = mpl.colors.Normalize(vmin=0, vmax=len(trg))
    
    rang = np.linspace(0.1, 1, num=len(trg))
    for i,col,lb in zip(np.arange(1,6), feats.columns, ['line count', 'verticle count', 'horizontal count', 'most x', 'most y']):
        ax = fig.add_subplot(3,2,i)
        ax.set_xlabel(lb)
        ax.set_ylabel('score')
        axs = ax.scatter(feats[col].tolist(), trg.score.tolist(), s=rang*30, c=cmap(rang))
    cb = mpl.colorbar.ColorbarBase(cax, cmap=cmap, norm=norm, orientation='horizontal')
    #cb.set_label('Some Units')
    fig.show()
    return

def evaluation(model, trFeats, trGnd, tsFeats, tsGnd):
    model.fit(trFeats, trGnd)
    predicted = model.predict(tsFeats)
    
    evHue = pd.DataFrame(index=np.arange(1,len(tsFeats)+1))
    evHue.insert(0, 'Ground', tsGnd)
    evHue.insert(1,'Predict', predicted)
    
    print '------------------------'
    print 'Pearson Evaluation'
    print evHue.corr(method='pearson')
    print '------------------------'
    print 'Spearman Evaluation'
    print evHue.corr(method='spearman')
    print '------------------------'
    print 'Kendall Evaluation'
    print evHue.corr(method='kendall')
    
    fig = mplt.figure(figsize=(14,7))
    fig.subplots_adjust( wspace=.3)
    ax = fig.add_subplot(1,2,1)
    #mplt.subplot(2,2,1)
    
    cmap = cm.PuBu
    cax = fig.add_axes([0.02, 0.1, 0.02, 0.8])
    norm = mpl.colors.Normalize(vmin=1, vmax=len(tsGnd))
    cb = mpl.colorbar.ColorbarBase(cax, cmap=cmap, norm=norm, orientation='vertical', ticks=np.arange(0, len(tsGnd)+1, 10))
    rang = np.linspace(.1, 1, len(tsGnd))
    
    fit = polyfit(evHue.Ground, evHue.Predict, 1)
    fit_fn = poly1d(fit) # fit_fn is now a function which takes in x and returns an estimate for y
    ax.plot(np.arange(0,11), fit_fn(np.arange(0,11)), '--g', lw=3, label='Fit Line')
    ax.scatter(evHue.Ground, evHue.Predict, s=40*rang, c=cmap(rang), label='Score')
    ax.legend(loc=2)
    ax.set_xlim( 0, 10)
    ax.set_ylim( 0, 10)
    ax.set_ylabel('Predicted')
    ax.set_xlabel('Ground Truth')
    
    ax = fig.add_subplot(1,2,2)
    ax.plot(np.arange(0,80), evHue.Ground, 'g-', label='Ground Truth')
    ax.plot(np.arange(0,80), evHue.Predict, 'b--', lw=2, label='Predicted')
    ax.legend()
    ax.grid(True)
    ax.set_ylabel('Score')
    ax.set_xlabel('Input Image')
    return predicted

def scatter3d(fig, gridIndex, x, y, z):
    ax = fig.add_subplot(gridIndex, projection='3d')
    xs = x
    ys = y
    zs = z
    ax.scatter3D(xs, ys, zs, c='r')

    #ax.set_xlim(-.1, .1)
    #ax.set_ylim(.1, -.1)
    #ax.set_zlim(-.1, .1)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    return
```

```python
radius = 5
img = ilTrain[23]
# Grab a little blobby image from the interwebs (it's a picture of a mouse, taken with the Microsoft Kinect)
scv_img = img
 
# Grab a grayscale layer of the image
img = scv_img.getNumpy()[:,:,0]
 
# Calculate the Euclidean distance transform of the image
# ( This will make a ridge in the center of the blob that will be the bones of our skeleton )
distance_img = ndimage.distance_transform_edt(img)
 
# Calculate the morphological laplacian
# ( this makes the ridge from the distance transform really pop out, it's a kind of an edge detection )
morph_laplace_img = ndimage.morphological_laplace(distance_img, (radius, radius))
 
# Threshold the Laplacian, and voila, we've got a nice ridge
skeleton = morph_laplace_img < morph_laplace_img.min()/2
 
# NOTE:
# To be really robust and hardcore, you'd probably want to parameterize the "skeleton" array somehow.
# What we get back is actually a mask, not a definition of a line or curve, but it wouldn't be too hard
# to take the (x,y) positions of non-zero points in the skeleton mask and fit a function through them.
# What the function is you'd like to fit depends on what you expect your skeleton to look like
# For a mouse, it's probably a cubic, which is pretty easy. 
# See http://www.scipy.org/Cookbook/FittingData for ideas on how to fit functions to data.
 
# Show the middle-ridge as a knock-out of the original image
#img[skeleton] = 0
#scv_img = Image(img)
scv_img.show()
ilTrain[23].show()
```
