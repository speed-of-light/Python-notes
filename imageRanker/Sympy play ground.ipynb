```python
from __future__ import division

import sympy as sym
import matplotlib.pylab as mplt
#from sympy import *

sym.init_printing(use_latex=True)
```

```python
xi, xj, yi, yj = sym.symbols('x_i, x_j, y_i, y_j')
#sx, sy = sym.symbols('sigma_x sigma_y')
rpos, rsize= sym.symbols('gamma_pos, gamma_size')

sx = sym.Float(1)
sy = sym.Float(1)
xj = yj = sym.Integer(0)
rpos = sym.Float(1/(sym.sqrt(2*pi)*1))

ex = rpos * sym.exp(((xi-xj)/sx)**2)
ey = sym.exp((yi-yj)**2/sy**2)
Apos = rpos * ex * ey #sym.symbols('A_ij^pos')
```

```python
sym.plotting.plot3d(ex*ey, (xi, -1, 1), (yi, -1, 1), size=100, xlabel='$x_i$', ylabel='$y_i$')
sym.plotting.plot3d(ex*ey, (xi, -3, 3), (yi, -3, 3), size=100, xlabel='$x_i$', ylabel='$y_i$')
```
