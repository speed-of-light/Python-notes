```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize(8,12))
ax = fig.add_subplot(3,1,1, projection='3d')
ax2 = fig.add_subplot(3,1,2, projection='3d')
ax3 = fig.add_subplot(3,1,3, projection='3d')

X = np.arange(-10, 10, 1)
Y = np.arange(-10, 10, 1)
X, Y = np.meshgrid(X, Y)

sx,sy = 1, 1
x, y = 0, 0
rpos = 1/(sqrt(2*pi)*.99)
Z1 = (np.exp(X-x)/sx)**2
Z2 = (np.exp(Y-y)/sy)**2
Z = rpos* Z1 * Z2
#Z = R
surf = ax.plot_surface(X, Y, Z1, rstride=2, cstride=2, cmap=cm.coolwarm,
        linewidth=1, antialiased=True)
ax2.plot_surface(X, Y, Z2, rstride=2, cstride=2, cmap=cm.coolwarm,
        linewidth=1, antialiased=True)
ax3.plot_surface(X, Y, Z, rstride=2, cstride=2, cmap=cm.coolwarm,
        linewidth=1, antialiased=True)
ax.set_ylim(10, -10)
#ax.set_zlim(.001, 4)

ax3.set_xlabel('$x_i$', fontsize=26)
ax3.set_ylabel('$y_i$', fontsize=26)
#ax.zaxis.set_major_locator(LinearLocator(10))
#ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```
