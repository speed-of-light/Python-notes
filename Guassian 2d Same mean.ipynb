```python
from pylab import *
%pylab inline
```

```python
#Default delta is large because that makes it fast, and it illustrates
# the correct registration between image and contours.
delta = 0.05
x = arange(-3.0, 3.001, delta)
y = arange(-3.0, 3.001, delta)
X, Y = meshgrid(x, y)

Z1 = bivariate_normal(X, Y, 1, 1, -1, 0)
Z2 = bivariate_normal(X, Y, 1, 1,  1, 1, .2)
#Z = (Z2 - Z1) * 10
Z = Z2 * 10
levels = arange(-3.0, 3.01, 0.2) 
title("Bivariate with mu = 1, mu = -1") 
norm = cm.colors.Normalize(vmax=abs(Z).max(), vmin=-abs(Z).max())
cmap = cm.PRGn
cset1 = contourf(X, Y, Z, levels,
                    cmap=cm.get_cmap(cmap, len(levels)-1),
                    norm=norm)
cset2 = contour(X, Y, Z, cset1.levels,
                    colors = 'r',
                    hold='on')
#figure(figsize=(5,5))

# We don't really need dashed contour lines to indicate negative
# regions, so let's turn them off.
#for c in cset2.collections:
#    c.set_linestyle('solid')
```

```python
figure(figsize=(10,10))
subplots_adjust(wspace=.2, hspace=.4)
rows = 5
for sx in range(3):
    for sy in range(3):
        Z1 = bivariate_normal(X, Y, 1.0, 1.0, 0.0, 0.0)
        Z2 = bivariate_normal(X, Y, 1.0 + sx, 1.0 + sy, 0.0, 0.0)
        Z = (Z2 - Z1) * 10
        # Boost the upper limit to avoid truncation errors.
        levels = arange(-3.0, 3.01, 0.2) 
        subplot(rows,3,sx*3+sy+1)
        title('(sx:%d, sy:%d)' % (sx, sy)) 
        norm = cm.colors.Normalize(vmax=abs(Z).max(), vmin=-abs(Z).max())
        cmap = cm.PRGn
        cset1 = contourf(X, Y, Z, levels,
                        cmap=cm.get_cmap(cmap, len(levels)-3),
                        norm=norm)
        cset2 = contour(X, Y, Z, cset1.levels,
                        colors = 'r',
                        hold='on')

for i in range(3):
    covxy = i * .2
    Z1 = bivariate_normal(X, Y, 1.0, 1.0, 0.0, 0.0, 0)
    Z2 = bivariate_normal(X, Y, 1.0, 1.0, 0.0, 0, covxy)
    Z = (Z2 - Z1) * 10
    # Boost the upper limit to avoid truncation errors.
    levels = arange(-3.0, 3.01, 0.2) 
    subplot(rows,3,10+i)
    title('(sxy = %d)' % (covxy)) 
    norm = cm.colors.Normalize(vmax=abs(Z).max(), vmin=-abs(Z).max())
    cmap = cm.PRGn
    cset1 = contourf(X, Y, Z, levels,
                    cmap=cm.get_cmap(cmap, len(levels)-1),
                    norm=norm)
    cset2 = contour(X, Y, Z, cset1.levels,
                    colors = 'r',
                    hold='on')

for i in range(3):
    covxy = i*.9
    Z1 = bivariate_normal(X, Y, 1.0, 1.0, 0.0, 0.0, 0)
    Z2 = bivariate_normal(X, Y, 3.0, 3.0, 0.0, 0, covxy)
    Z = (Z2 - Z1) * 10
    # Boost the upper limit to avoid truncation errors.
    levels = arange(-3.0, 3.01, 0.2) 
    subplot(rows,3,13+i)
    title('(sxy = %d)' % (covxy)) 
    norm = cm.colors.Normalize(vmax=abs(Z).max(), vmin=-abs(Z).max())
    cmap = cm.PRGn
    cset1 = contourf(X, Y, Z, levels,
                    cmap=cm.get_cmap(cmap, len(levels)-1),
                    norm=norm)
    cset2 = contour(X, Y, Z, cset1.levels,
                    colors = 'r',
                    hold='on')
```

```python
figure(figsize=(10,10))
subplots_adjust(wspace=.2, hspace=.4)
rows = 3
for sx in range(3):
    for sy in range(3):
        Z1 = bivariate_normal(X, Y, 1, 1, 3, 0.0)
        Z2 = bivariate_normal(X, Y, 1.0 + sx, 1.0 + sy, -3, 0.0)
        Z = (Z2 - Z1) * 10
        # Boost the upper limit to avoid truncation errors.
        levels = arange(-3.0, 3.01, 0.2) 
        subplot(rows,3,sx*3+sy+1)
        title('(sx:%d, sy:%d)' % (sx, sy)) 
        norm = cm.colors.Normalize(vmax=abs(Z).max(), vmin=-abs(Z).max())
        cmap = cm.PRGn
        cset1 = contourf(X, Y, Z, levels,
                        cmap=cm.get_cmap(cmap, len(levels)-3),
                        norm=norm)
        cset2 = contour(X, Y, Z, cset1.levels,
                        colors = 'r',
                        hold='on')
```

```python
# It is easier here to make a separate call to contour than
# to set up an array of colors and linewidths.
# We are making a thick green line as a zero contour.
# Specify the zero level as a tuple with only 0 in it.
cset3 = contour(X, Y, Z, (0,),
                colors = 'g',
                linewidths = 2,
                hold='on')
title('Filled contours')
colorbar(cset1)
#hot()


subplot(2,2,2)
extent = (-3,3,-3,3)
imshow(Z, extent=extent, cmap=cmap, norm=norm)
v = axis()
contour(Z, levels, hold='on', colors = 'k',
        origin='upper', extent=extent)
axis(v)
title("Image, origin 'upper'")

subplot(2,2,3)

imshow(Z, origin='lower', extent=extent, cmap=cmap, norm=norm)
v = axis()
contour(Z, levels, hold='on', colors = 'k',
        origin='lower', extent=extent)
axis(v)
title("Image, origin 'lower'")

```

```python
subplot(2,2,4)

# We will use the interpolation "nearest" here to show the actual
# image pixels.
# Note that the contour lines don't extend to the edge of the box.
# This is intentional. The Z values are defined at the center of each
# image pixel (each color block on the following subplot), so the
# domain that is contoured does not extend beyond these pixel centers.
im = imshow(Z, interpolation='nearest', extent=extent, cmap=cmap, norm=norm)
v = axis()
contour(Z, levels, hold='on', colors = 'k',
        origin='image', extent=extent)
axis(v)
ylim = get(gca(), 'ylim')
setp(gca(), ylim=ylim[::-1])
title("Image, origin from rc, reversed y-axis")
colorbar(im)

show()
```
